```r
data("iris")
head(iris)
```

```r
mytheme<-theme(
    axis.title = element_text(size = 18),      
    axis.text = element_text(size = 16),       
    plot.title = element_text(size = 18),      
    legend.text = element_text(size = 16),     
    legend.title = element_text(size = 18)     
  )
theme_set(mytheme)
```

```r
library(ggplot2)
options(repr.plot.width=12, repr.plot.height=9)
ggplot(data=iris, aes(x=Petal.Length, fill=Species)) +            
                geom_density(alpha=.3)
```

```r
ggplot(data=iris, aes(x=Petal.Length, fill=Species)) + 
  geom_histogram(alpha=0.8, position="identity", bins=30)
```

```r
ggplot(data=iris,aes(x=Species,y=Petal.Length)) + 
  geom_boxplot()+
  geom_point(position = "jitter",colour="black")
```

```r
ggplot(data=iris,aes(x=Sepal.Width,y=Sepal.Length,colour=Species,shape=Species))+geom_point(size=5)
```

```r
iris_summary <- iris %>%
  group_by(Species) %>%
  summarise(
    Petal.Length = sum(Petal.Length),
    Petal.Width = sum(Petal.Width),
    Sepal.Length = sum(Sepal.Length),
    Sepal.Width = sum(Sepal.Width),
    .groups = 'drop' # drop grouping for further operations
  )

iris_long <- iris_summary %>%
  pivot_longer(
    cols = -Species, # Select all columns except Species
    names_to = "variable",
    values_to = "value"
  )

ggplot(iris_long, aes(x = Species, y = value, fill = variable)) +
  geom_bar(stat = "identity", position = "stack") 
```

```r
ggplot(iris_long,aes(x=Species,y=value,fill=variable))+
geom_bar(stat="identity",position="dodge")
```

```r
ggplot(data=iris,aes(x=Sepal.Length,y=Sepal.Width,colour=Species))+
facet_grid(Species~.)+
geom_point()
```
